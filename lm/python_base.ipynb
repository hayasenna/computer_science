## Переменные

`Переменной в Python называется имя, которое ссылается на объект в памяти.`

То есть переменная это не ячейка со значением, как, например в математике, а ссылка.  
`И В НЕКОТРЫХ МОМЕНТАХ С ЭТИМ НУЖНО БЫТЬ ОЧЕНЬ АККУРАТНЫМ!` Но, сегодня не об этом.


Тип переменной в python определяется `автоматически`: данное явление называется дианмической типизацией. Тип определяется сразу после присовения переменной значения. У такой типизациии есть как и свои плюсы, так и минусы, но сегодня не об этом.

Присваивание переменной значения определяется знаком `=`. Например:



```python
x = 12 # где "x" - название переменной, а "12" - значение
```

## Операторы

`Оператор — это специальный символ или ключевое слово, которое даёт инструкцию компьютеру выполнить определённое действие над одним или несколькими объектами.`

Операторы в Python можно поделить на несколько групп:
* Арифметические - для выполнения мат. операций
    * Сложение: `+`
    * Вычитание: `-`
    * Умножение: `*`
    * Деление: `/`
    * Целочисленное деление: `//`
    * Остаток от деления: `%`
    * Возведение в степень: `**`
* Сравнения - для оценки логических выражений, результат - True/False
    * Равно: `==`
    * Не равно: `!=`
    * Больше: `>`
    * Меньше: `<`
    * Больше или равно: `>=`
    * Меньше или равно: `<=`
* Логические - для работы с булевыми значениями, результат - True/False
    * И: `and`
    * ИЛИ: `or`
    * НЕ/Отрицание: `not`
* Присваивания  
    **Здесь x - исходная переменная, n - какое-то число**
    * x = n: `тут все как есть`
    * x = x + n: `x += n`
    * x = x - n: `x -= n`
    * x = x * n: `x *= n`
    * x = x + n: `x += n`
    * x = x / n: `x /= n`
    * x = x // n: `x //= n`
    * x = x % n: `x %= n`
    * x = x ** n: `x **= n`
* Принадлежности - проверка принадлежности чего-то к чему-то, результат - True/False
    * x является частью n: `x in n`
    * x не является частью n: `x not in n`
* Идентичности - проверка на один и тот же экземпляр объекта, результат - True/False
    * x является частью n: `x is n`
    * x не является частью n: `x is not n`
* Побитовые
    * И: `&`
    * ИЛИ: `|`
    * XOR/Умножение: `^`
    * НЕ: `~`
    * Сдвиг влево: `<<`
    * Сдвиг вправо: `>>`


## Типы данных

`Тип данных — это категория, определяющая возможные значения, которые может принимать объект, а также операции, которые можно над этими значениями выполнять`

Типы данных в Python можно очень условно делить на группы. Начнем с разделения на 5 больших:
1. Числовые
    * `int` - целое чилсло
    * `float` - вещественное чило
    * `complex` - комплексное число
2. Строковые
    * `str` - строка
3. Массивы/коллекции
    * `list` - список 
    * `dict` - словарь 
    * `tuple` - кортеж 
    * `set` - множество 
    * `frozenset` - Замороженное множество
4. Логические 
    * `bool` - булевый тип
5. Специальный: 
    * `NoneType` - None/"Ничего"

```python
""" 
ВСЕ ВЫВОДЫ В КОНСОЛЬ ПРОНУМЕРОВАНЫ! НЕ ТЕРЯЙТЕСЬ!
"""

# Числовые типы данных

integer = 5
float_num = 45.3
complex_num = (-1)**(1/2)

print(f"1. Value: {integer}, type: {type(integer)}")
print(f"2. Value: {float_num}, type: {type(float_num)}")
print(f"3. Value: {complex_num}, type: {type(complex_num)}")

# Строки

char = 'a'
string = 'aboba' # В обоих случаях можно было указать еще и с двойными ковычками, тут без разницы: char = "a"
print(f"4. Value: {char}, type: {type(char)}")
print(f"5. Value: {string}, type: {type(string)}")

# Массивы

nums_list = [1, 2, 3, 4, 5, 6]
nums_dict = {"a" : 7, "b" : 8, "c" : 9} # Хранит ключ (в данном случае, например "a") и значение (в данном случае, например для "a", 7)
nums_tuple = (10, 247, 23, 43)
nums_set = {22, 34123, 234} # Хранит только уникальные элементы (работает так же, как в математике). Например: set([1,1,2,3]) вернет множество {1,2,3} (да, множество модно хадать не только фигурлными скобками.
nums_frozentset = frozenset("aboba228")

print(f"6. Value: {nums_list}, type: {type(nums_list)}")
print(f"7. Value: {nums_dict}, type: {type(nums_dict)}")
print(f"8 Value: {nums_tuple}, type: {type(nums_tuple)}")
print(f"9. Value: {nums_set}, type: {type(nums_set)}")
print(f"10. Value: {nums_frozentset}, type: {type(nums_frozentset)}")


# Логичексий тип

logical = True
sentence = (1 == 0)

print(f"11. Value: {logical}, type: {type(logical)}")
print(f"12. Value: {sentence}, type: {type(sentence)}")

# NoneType

a = None

print(f"13. Value: {a}, type: {type(a)}")
```

## Условные конструкции
`Условная конструкция — это конструкция, позволяющая выполнить блок кода только при выполнении определённого условия.`  

Работает это так:  
Есть входное условие `if`. Еслив коде есть невложенный `if` - он обязательно проверится. То есть:

```python
a = 123
b = 'chief_kief'

# последоваетльное выполение

if a == 122: # Зашли в первый (условие не выполняется)
    print(b)
if type(b) == str: # Зашли во второй (условие выполняется)
    print(b)
if a == 123: # Зашли в третий (условие выполняется)
    print(a)
```

На случай же невыполнения `if`, можно предусмотреть `elif` (else-if) и `else`. Но если `if` выполнится, то мы даже смотреть не пойдем, что там в следующих за ним кострукциях. Например:

```python
a = "bimbambam"

if not(a):
    print("not a") # условие не выполнилось -> идем к следующему
elif len(a) > 3:
    print(a) # условие выполнилось, можно заканчивать и else уже не смотреть
else:
    pass
```

Ну и на закрепление:

```python
"""
Синтаксис!!!!!!
if 'logical sentence':
    ... some code 
elif 'logical sentence':
    ... some code 
else:
    ... some code 
"""


# База

a = True
if a:
    print("Boolean variable 'a' is always 'True'")

if 15 > 110: # скипаем блок потому что: 15 > 110 <=> False
    print("False")
elif 15 == 101: # тоже скип
    print("False")
else:
    print("True")


# По-веселее

logical_sentences = [0 == 1, 'abc' == 'bca', len('abc') == len('bca'), 1239 >= 1000000, 'ab' in 'abcds']
counter = 0

for i in range(len(logical_sentences)):
    if logical_sentences[i]:
        counter += 1
        print(f"Sentence №{i} is {logical_sentences[i]}") # Наопмниаю, что индексация в массивах - с нуля, т.е первый элемент имеет индекс 0

print(f"{counter} sentences were True")
    
```

## Циклы

`Циклом, в любом языке программирования, называется конструкция позволяющая выполнить код находящийся внутри него некторое количеcтво раз.`  

Каждое отдельное повторение называется `итерацией`.

Всего в python существует два цикла: `for` и `while`.  
`Цикл for либо работает с коллекциями, а в частности с их элеменатми, либо в ситуациях когда заранее известно количество итераций:`

```python
"""
Синтаксис!!!!!!
for 'iterator' in 'collection':
    ... some code 
"""

# Например

for i in range(5): # ф-ция range генерирует последовательность, в нашем случае [0;5) с шагом = 1. Все параметры можно менять.
    print(i) # подробнее про range: https://docs.python.org/3/library/functions.html#func-range

# Или

arr = ['apple', 'microsoft', 'google', 'syoma-tester-inc']

# Проходимся по всем элементам массива и выводим их.
for elem in arr:
    print(elem)

# Напомню, что внутрь цикла можно вставить более функциональный код:

hash_table = dict() # пустой словарь

for elem in arr: 
    hash_table[len(elem)] = elem # проходимся по массивц строк и: 

# 1. записываем длины строк в качестве ключей
# 2. А в качестве значей записываем ссответсвующие строки


sorted_dict = dict(sorted(hash_table.items())) # сортируем и конвертируем список, отданный ф-цией sorted обратно в словарь
print(sorted_dict)
```

`Цикл while работает от заданного условия: пока условие в его основе верно - он работает.`  
`Используется в случаях, когда количсевто итераций заранее неизвестно.`

```python
"""
Синтаксис!!!!!!
while 'logical sentence':
    ... some code 
"""

i = 0
while i < 10:
    print(i)
    i += 1 # не забывем увеличивать i на 1, чтобы не было бесконечного цикла
print('finish'+'\n')



# НОК двух чисел:

a = 15
b = 15
ans = 1 # переменная для ответа


while True:
    if ans % a == 0 and ans % b == 0:
        break # про это чуть позже
    ans = ans + 1
print(ans) 
```

### Операторы циклов

В Python существует 2 оператора циклов: `break` и `continue`.

`Оператор break прерывает выполнение цикла, если выполняется подводящее к нему условие.` Например:

```python
nums = [1, 2, 3, 5, 6, 2, 5, 37.37, 66, 38]

for i in range(len(nums)):
    if type(nums[i]) == float: # Выходим, если видим вещественное число
        print(f"Here's a breakpoint, because the next element is {nums[i]}, which type is 'float'\n")
        break
    print(nums[i])


# Или 

i = 0

while i < len(nums):
    if nums[i] > 50:
        break
    print(nums[i])
    i += 1
```

`Оператор continue пропускает итерацию, на которой выполняется подводящее к нему условие.` Например:

```python
messages = ['apple', 'banana', 'ABOBA', 'dragonfruit']

for char in messages:
    if char == 'ABOBA':
        print("Skip skip skip skip") # Скипаем абобу
        continue
    print(char)


# Это чисто ради примера с while, но так лучше не делать, можно накосячить с обновлением переменной i (инкрементом i += 1)
i = -10
while i < 11:
    if i == 0:
        print("Skip 0")
        i += 1
        continue
    print(i)
    i += 1

# Лучше вот так (без continue (замена на else)):

i = -10
while i < 11:
    if i == 0:
        print("Skip 0")
    else:
        print(i)
    i += 1
```

Не могу назвать это оператором, но вещь интересная, поэтому скажу.  
Речь о конструкции `for .. else`:

```python
# Работает
numbers = [1, 3, 5, 7, 9]

for num in numbers:
    if num % 2 == 0:
        print("Найдено чётное число:", num)
        break
else:
    print("Чётных чисел не найдено.")
```

Суть в том, что блок `else` выполняется только в случае полного завершения работы цикла.

```python
# Не работает
numbers = [1, 3, 5, 6, 7, 9]

for num in numbers:
    if num % 2 == 0:
        print("Найдено чётное число:", num)
        break
else:
    print("Чётных чисел не найдено.")
```

## Функции

`Функция — это именованный блок кода, который выполняет определённую задачу. Функции помогают структурировать программу, делают код более читаемым и позволяют использовать один и тот же код многократно.` 

```python

"""
Синтаксис!!!!!!
def 'function_name'(arguments):
    (return)
"""

# Хорошим тоном является типизация всего (аргументов и выходных значений), что я и сделал.
# Функция может быть записана так:
# def f(x):
#     return x**2
# Но если я четко знаю, чего я хочу от функции, то я типизирую:

def f(x: int) -> int: # типо парабола
    return x**2 # слово return - возвращает значение из функции. если функция ничего не возвращает (такое может быть), то слова return может не быть вовсе.

for i in range(-5, 6):
    print(f(i))

#  ДАЛЕЕ - ЛЕГЕНДА
```

Как понять нужен `return` или нет? Все просто - если вы планируете использовать в коде значение, полученное через функцию - нужно его вернуть (return).

```python
import time
import sys
from typing import Any, Optional, Tuple, Literal, Iterable


# Функция ничего не возвращает (None), а просто выводит текст на экран. 


def karaoke(*args) -> None: # *args - это набор позиционных аргументов (нам это пока не надо)

    timed_words = [
        (0.0, 'В ладони'),
        (1.0, 'ЛАБУБА'),
        (1.9, 'Во рту'),
        (2.5, 'ХУБА-БУБА'),
        (3.6, 'Целую тебя в губы'),
        (5.4, 'Сердце моё'),
        (6.1, 'ТУК-ТУК-ТУК-ТУК'),
        (7.2, 'В ладони'),
        (8.2, 'ЛАБУБА'),
        (9.0, 'Во рту'),
        (9.7, 'ХУБА-БУБА'),
        (11.0, 'Целую тебя в губы'),
        (12.4, 'Сердце моё послушай'),
        (13.2, '(Чё?)')
    ]

    start_time = time.time()

    for timestamp, word in timed_words:
        now = time.time()
        sleep_time = timestamp - (now - start_time)
        if sleep_time > 0:
            time.sleep(sleep_time)
        sys.stdout.flush()
        print(word)

    print("\n")
    for arg in args:
        time.sleep(1)
        print(arg + "\n")



karaoke('AAAAAAAAAAAAAAAAA!!!!!', 'Удоли текток..', 'blablabla', "blebleble", "blublublu")
```

Вот более полезная штука - `бинарный поиск`:

```python
# Принимает на вход коллекцию и значение, позицию (индекс) которого мы хотим найти. Работет только в отсортированном массиве!!!
# Если интересно: https://habr.com/ru/articles/783848/

def bin_search(array: Iterable, target: int) -> Optional[int]:
    low = 0
    high = len(array) - 1

    while low <= high:
        mid = (low + high) // 2
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            high = mid - 1
        else:
            low = mid + 1
    return None


my_list = [1, 2, 3, 3, 4, 5, 6, 7, 49, 330, 1000, 1115]

print(bin_search(my_list, 6))
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np


# Задаем функцию

def plot_3d(
    x_data: Any, y_data: Any, z_data: Any,
    sides: Tuple[int], px: float, trace = None, 
    x_title: Optional[str] = None, y_title: Optional[str] = None, z_title: Optional[str] = None,
    mode: Literal['scatter', 'heatmap', 'contour_lines_plot'] = 'scatter', colored: bool = False,
    color1: Optional[str] = None,
    color2: Optional[str] = None,
    color3: Optional[str] = None,
    range1: Optional[int] = None,
    range2: Optional[int] = None,
    range3: Optional[int] = None
    ) -> "plot_3d":
    


    fig = plt.figure(figsize=sides, dpi=px)
    ax = fig.add_subplot(111, projection="3d")


    if mode == 'heatmap':
        surf = ax.plot_surface(x_data, y_data, z_data, cmap=cm.coolwarm, linewidth=0, antialiased=False)
        fig.colorbar(surf, shrink=0.5, aspect=5)
    

    elif mode == 'scatter':
        if colored:
            if range1:
                colors = [color1 for _ in range(range1)]
                if range2:
                    colors = np.concatenate((colors, [color2 for _ in range(range2)]))
                    if range3:
                        colors = np.concatenate((colors, [color3 for _ in range(range3)]))

        ax.scatter(x_data, y_data, z_data, c = colors)
    

    elif mode == 'contour_lines_plot':
        ax.plot_surface(x_data, y_data, z_data, alpha=0.6)
        ax.contour(x_data, y_data, z_data, offset=z_data.min(), cmap=cm.coolwarm)

        if trace is not None:
            ax.plot(trace[:, 0], trace[:, 1], trace[:, 2], "o-", color="red", linewidth=2)
    
            ax.set_xlim(np.min(x_data), np.max(x_data))
            ax.set_ylim(np.min(y_data), np.max(y_data))
            ax.set_zlim(np.min(z_data), np.max(z_data))
    
    else:
        raise ValueError("Incorrect mode chosen!")

    ax.set_xlabel(x_title)
    ax.set_ylabel(y_title)
    ax.set_zlabel(z_title)


    plt.show()
```

```python
# Юзаем

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x_data, y_data = np.meshgrid(x, y)

# Поверхность z = sin(sqrt(x^2 + y^2))
z_data = np.sin(np.sqrt(x_data**2 + y_data**2))

# Добавим трассер (путь), например, вдоль x = y
trace_points = np.linspace(-5, 5, 50)
trace = np.array([[v, v] for v in trace_points])  # x = y
trace_z = np.sin(np.sqrt(trace[:, 0]**2 + trace[:, 1]**2))
trace_full = np.column_stack((trace, trace_z))

plot_3d(
    x_data=x_data,
    y_data=y_data,
    z_data=z_data,
    x_title='X Axis',
    y_title='Y Axis',
    z_title='Z = sin(sqrt(x^2 + y^2))',
    sides=(10, 10),
    px=100,
    # trace=trace_full,
    mode='contour_lines_plot'
)
```
